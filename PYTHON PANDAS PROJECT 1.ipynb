RELATIONSHIP ANALYSIS

```python
KCQT= pd.read_csv("KCQT.csv")
KCQT
```

CORELLATION MATRIX

```python
# I have generated a correlation matrix. The matrix will help summarize the relationship between the variables by checking for the coefficient values indicated.
corelation=KCQT.corr() 
```

```python
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

**Negative coefficients indicate a negative relationship between the associated variables; where one variable tends to decrease with an increase of the other whereas a positive coefficient would mean one variable tends to increase with an increase of the other.

To show the relationship between each 2 variables more specifically, i've used pairplot 
the distribution in the plots also indicates the relationship between the variables. 

```python
sns.pairplot(KCQT)
```

```python
#relationship plot 
sns.relplot(x="average_max_temp", y="average_precipitation", hue='actual_precipitation', data=KCQT)
```

*The scatter plot data display slops downward left to right, showing a negative relationship betweeen average_max_temp and average_precipitation. 
aveage_precipitation tends to decrease as average_max_temperature increases  

the same finding is backed by the correlation matrix diagram generated above, that indicates a negative coefficient between the 2 variables as seen at their intersection.
